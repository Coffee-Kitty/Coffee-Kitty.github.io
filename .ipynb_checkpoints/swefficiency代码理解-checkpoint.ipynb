```python
import ast

code = "def f(x: int) -> str:\n    return str(x)\n"
tree = ast.parse(code)

for node in ast.walk(tree):
    if isinstance(node, ast.arg) and node.annotation:
        print(node.annotation.lineno, node.annotation.col_offset)
```

```python
import ast
import asttokens

code = "result = (1 + 2) * 3"
# 1. 标记代码
atok = asttokens.ASTTokens(code, parse=True)

# 2. 遍历 AST
for node in ast.walk(atok.tree):
    if isinstance(node, ast.BinOp):
        # 3. 获取该节点在原代码中的样子
        print(f"节点: {type(node).__name__}, 代码: {atok.get_text(node)}")
# 输出: 节点: BinOp, 代码: (1 + 2) * 3
```

```python
node = atok.tree.body[0].value # 获取表达式节点
start, end = atok.get_text_range(node)
print(f"位置: {start}到{end}, 内容: {code[start:end]}")
```

```python
import jedi

code = """
import math
x = math.sqrt(4)
"""

script = jedi.Script(code, path="example.py")
names = script.get_names(all_scopes=True, definitions=False, references=True)

for n in names:
    print(n.name, n.type, n.line, n.column)
```

```python
import jedi

code = """
from math import sqrt

def foo(x):
    return sqrt(x)

y = foo(9)
"""

script = jedi.Script(code, path="example.py")

print("=== get_names ===")
names = script.get_names(all_scopes=True, definitions=False, references=True)
for n in names:
    print(
        f"name={n.name!r}, type={n.type!r}, line={n.line}, column={n.column}"
    )

print("\n=== goto from references named 'sqrt' ===")
for n in names:
    if n.name == "sqrt":
        defs = n.goto(follow_imports=True, follow_builtin_imports=False)
        print(f"\nReference: {n.name!r} at line {n.line}, col {n.column}")
        for d in defs:
            print(
                f"  -> goto name={d.name!r}, type={d.type!r}, "
                f"module_path={d.module_path}, line={d.line}, column={d.column}"
            )
```

```python
code1 = "runtimes = timeit.repeat(workload, number=N, repeat=R, setup=setup)"
code2 = "runtimes = repeat(workload, number=N, repeat=R, setup=setup)"
tree = ast.parse(code2)
for node in ast.walk(tree):
    if (
        isinstance(node,ast.Assign)
        and isinstance(node.value,ast.Call)
    ):
        call = node.value
        func,args,keywords = call.func, call.args, call.keywords
        print(type(call.func))
        print(len(call.args))
        print(call.args[0].id)
        print(len(call.keywords))
        
        
        if isinstance(call.func, ast.Attribute):
            print(call.func.attr)
            print(call.func.value.id)
            
        if isinstance(call.func, ast.Name):
            print(call.func.id)
```
