```python
from rich.progress import Progress, SpinnerColumn, BarColumn, TextColumn, TimeRemainingColumn
from rich.live import Live
from rich.panel import Panel
from rich.table import Table
from rich.console import Console
```

```python
from time import sleep
from rich.progress import Progress, SpinnerColumn, BarColumn, TextColumn, TimeRemainingColumn
from rich.live import Live
from rich.panel import Panel
from rich.table import Table

progress = Progress(
    SpinnerColumn(),
    TextColumn("{task.description}"),
    BarColumn(),
    TextColumn("{task.percentage:>3.0f}%"),
    TimeRemainingColumn(),
)

status = Progress(
    SpinnerColumn(),
    TextColumn("{task.description}"),
)

layout = Table.grid()
layout.add_row(Panel(progress, title="Main Progress"))
layout.add_row(status)

with Live(layout, refresh_per_second=10):
    task1 = progress.add_task("Downloading", total=100)
    task2 = status.add_task("Preparing...", total=None)

    for i in range(100):
        progress.update(task1, advance=1)
        status.update(task2, description=f"Step {i}")
        sleep(0.05)
```

```python
import time
from rich.progress import Progress, SpinnerColumn, BarColumn, TextColumn, TimeRemainingColumn


def do_work(i: int) -> int:
    time.sleep(0.2)
    return i * 2


with Progress(
    SpinnerColumn(),
    TextColumn("{task.description}"),
    BarColumn(),
    TextColumn("{task.completed}/{task.total}"),
    TimeRemainingColumn(),
) as progress:
    task = progress.add_task("Running", total=5)

    for i in range(5):
        result = do_work(i)
        progress.update(task, advance=1, description=f"Running (last={result})")
```
