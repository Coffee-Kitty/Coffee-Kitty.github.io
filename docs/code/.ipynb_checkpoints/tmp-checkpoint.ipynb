```python
a = [1,2,3]
b = [4,5,6]
print(list(zip(a,b)))
""" 
[(1, 4), (2, 5), (3, 6)]
"""
print(*zip(a,b))
print(list(zip( *zip(a,b) )))
""" 
(1, 4) (2, 5) (3, 6)
[(1, 2, 3), (4, 5, 6)]
"""
```

```python
y = map(lambda x:x+1, [1,2,3])
print(list(y))
""" 
[2, 3, 4]
"""
z = map(lambda x,y:x+y, [1,2,3],[4,5,6])
print(list(z))
"""
[5, 7, 9]
"""
```

```python
class A:
    def __new__(cls, *args, **kwargs):
        instance = super().__new__(cls, *args, **kwargs)
        print("--- new ----")
        return instance

    def __init__(self):
        print('--- init -----')

    def __del__(self):
        print('--- del ---')

a = A()
del a
"""  
--- new ----
--- init -----
--- del ---
"""
```

```python
class Person:
    def __init__(self, name):
        self.name = name
    def __str__(self):
        return f'I am {self.name}'
    def __repr__(self):
        return f'{self.name}'


p1 = Person('xinlin')
print(str(p1))

print(repr(p1))

print(p1)
""" 
'I am xinlin'
'xinlin'
I am xinlin
"""
```
